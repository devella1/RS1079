#### Function for Calculating the actual and accurate displacement  

```python
from math import radians, cos,sin, asin,sqrt

def distance(lon1, lon2, lat1 , lat2):    
    
    lon1 =radians(lon1)
    lon2 =radians(lon2)
    lat1 =radians(lat1)
    lat2 =radians(lat2)
    
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    
    r = 6371
    return(round (c * r, 2))
    
```

```python
longour, latour = -73.980658,	40.733873	
bufd = 0.6
r = 6371;pie = 3.142
lonrag = 180*bufd/(pie*r*cos(radians(longour)))
latrag = 180*bufd/(pie*r)
```

### Function for checking whether a point lyies in the a probabale range or not

```python

def inrange(lon,  lat ,  bufd):
    return (abs(latour - lat)<latrag) and (abs(longour - lon)<lonrag)
```

```python
d = []; ds = []
for i in range(df.shape[0]):
    k = distance(df['longitude_of_pickup'][i],longour,
                      df['latitude_of_dropoff'][i],
                      latour)
    kk = inrange(df['longitude_of_pickup'][i],
                      df['latitude_of_dropoff'][i],                      
                      bufd)
    d.append([k< bufd, kk]);
```

```python
print("True Positives:",d.count([True,True]),  "\tValues that are within the radius and within the range. ")
print("Missed Values :",d.count([True,False]), "\tValues that are within the radius but not within the range. ")
print("Extra Values  :",d.count([False,True]), "\tValues that are within the range but not within the range. ")
print("True Negatives:",d.count([False,False]),"\tValues that are neither within the radius nor within the range. ")
```
